```python
import pandas as pd
def extract_job_info(job_element):
    response1=requests.get(job_element.find_all('a',class_="card-footer-item")[1]['href'])
    soup1=BeautifulSoup(response1.content,'html.parser')
    description=soup1.find('div',class_="content").find_all('p')[0].text


    job_data={'title':job_element.find('h2',class_="title is-5").text ,
              'company':job_element.find('h3',class_="subtitle is-6 company").text ,
              'lacation':job_element.find('p',class_="location").text.strip() ,
              'time':job_element.find('time').text ,
              'description':description }
    return job_data


job_list=[]
for i in range(len(job_elements)):
    job_data=extract_job_info(job_elements[i])
    job_list.append(job_data)


df=pd.DataFrame(job_list)
df.to_csv('jobs.csv',index=False)




```
